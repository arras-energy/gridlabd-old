```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
N = 2000

xm = rnd.random()+1
xv = rnd.random()
x = rnd.normal(xm,xv,N)

ym = rnd.random()+2
yv = rnd.random()
y = rnd.normal(ym,yv,N)
```

```python
z = x+y
zm = z.mean()
zv = z.std()
```

```python
plt.figure(1,figsize=(12,5))
plt.hist(x,bins=int(N/50),density=True,label=f'x=N({xm:.2f},{xv:.2f})',color='red')
plt.hist(y,bins=int(N/50),density=True,label=f'y=N({ym:.2f},{yv:.2f})',color='blue')
plt.hist(z,bins=int(N/50),density=True,label=f'z=x+y=N({zm:.2f},{zv:.2f})',color='purple')
da = xm-4*xv
db = zm+4*zv
dr = 0.01
d = np.arange(da,db,dr)
def f(d,m,v): return np.array(list(map(lambda t:1.0/sqrt(2*np.pi)/v*exp(-((t-m)*(t-m))/(2*v*v)),d)))
plt.plot(d,f(d,xm,xv),color='red')
plt.plot(d,f(d,ym,yv),color='blue')
plt.plot(d,f(d,zm,zv),color='purple')
plt.legend()
plt.grid()
plt.show()
```

```python
print(f'N =',N)
print(f'x = N({xm:.2f},{xv:0.2f})')
print(f'y = N({ym:.2f},{yv:0.2f})')
print(f'z_obs = N({zm:.2f},{zv:0.2f})')
print(f'z_exp = N({xm+ym:.2f},{sqrt(xv*xv+yv*yv):.2f})')
```
